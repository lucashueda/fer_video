```python
import dlib
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
p = "shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
imagem = cv2.imread('perfil.jpg')

# frame = cv2.flip(imagem,90) 

imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem)
```

```python
rects = detector(imagem, 0)
rects
```

```python
%%time
for k,d in enumerate(rects): #For each detected face  
    shape = predictor(imagem, d) #Get coordinates
    for i in range(1,68): #There are 68 landmark points on each face
        cv2.circle(imagem, (shape.part(i).x, shape.part(i).y), 3, (0,255,0), thickness=-1) #For each point, draw a red circle with thickness2 on the original frame

            
plt.figure(figsize=(10,10))
plt.imshow(imagem) #Display the frame
plt.axis('off')
# plt.savefig('/content/drive/My Drive/Jupyter COLAB/Affective computing/Final project/Dlib_landmarks/landmarks.png')
```
